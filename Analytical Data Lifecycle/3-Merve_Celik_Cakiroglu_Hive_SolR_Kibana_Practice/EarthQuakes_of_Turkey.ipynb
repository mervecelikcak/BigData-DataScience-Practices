Combinaremos las columnas de 'Date of Occurrence' y 'Time of Occurrence' para crear una columna única que contenga la fecha y la hora juntas. Esto facilitará el análisis de los datos.

```python
# Concatenate the columns 'Date of Occurrence' and 'Time of Occurrence' into a new column 'Date_Time'
quake['Date_Time'] = pd.to_datetime(quake['Date of Occurrence'].str.cat(quake['Time of Occurrence']), format='%Y.%m.%d%H:%M:%S.%f')

# Drop the columns 'Date of Occurrence' and 'Time of Occurrence'
quake.drop(['Date of Occurrence', 'Time of Occurrence'], axis=1, inplace=True)

# Check the data types of the columns again
print("Types of the columns:")
print(quake.dtypes)
print()

# Check the number of rows and columns
print("Number of rows and columns:")
print(quake.shape)
```

Hemos cambiado los tipos de columnas. Ahora, debemos comprobar si hay valores perdidos.

```python
# Check the missing values
print("Missing values:")
print(quake.isnull().sum())
print("Total missing values:", quake.isnull().sum().sum())
```

Remplazamos los valores perdidos en la columna Mw (Magnitud de momento) con 0. Esto se debe a que no queremos que cambie el valor de la columna xM (El mayor valor entre los valores de magnitud dados). Si remplazamos con la media o la moda de la columna con los valores perdidos, los valores de xM pueden cambiar el valor, lo que puede dar lugar a resultados incorrectos.

**BOXPLOT**

**El diagrama de caja** y bigotes es una herramienta gráfica que representa la distribución de datos. Los elementos principales del diagrama son el cuadro, la línea dentro de la caja y los bigotes.

**El cuadro** representa el rango intercuartílico (IQR), que es el rango entre el primer cuartil (Q1) y el tercer cuartil (Q3). La longitud del cuadro indica la dispersión del 50% medio de los datos.

**La línea dentro de la caja** representa la mediana (Q2) del conjunto de datos. Da una idea de la tendencia central de los datos.

**Los bigotes** se extienden desde los bordes del cuadro hasta los valores mínimo y máximo dentro de un rango determinado. Los valores atípicos son puntos de datos que se encuentran fuera de los bigotes.

**En este caso, no hay necesidad de eliminar los valores atípicos porque los datos están dentro de un rango razonable. La profundidad de un terremoto puede variar desde 0 hasta 200 km y la magnitud de un terremoto puede variar desde 0 hasta 8. Estos rangos son lo suficientemente amplios como para incluir los valores atípicos sin que se consideren errores.**

```python
# Plot a boxplot
fig, axes = plt.subplots(9, 1, figsize=(20, 15))
sns.boxplot(x=quake['Latitude'], ax=axes[0], color="paleturquoise")
sns.boxplot(x=quake['Longtitude'], ax=axes[1], color="lemonchiffon")
sns.boxplot(x=quake['Depth'], ax=axes[2], color="lightcoral")
sns.boxplot(x=quake['xM'], ax=axes[3], color="violet")
sns.boxplot(x=quake['MD'], ax=axes[4], color="lightgreen")
sns.boxplot(x=quake['ML'], ax=axes[5], color="lightpink")
sns.boxplot(x=quake['Mw'], ax=axes[6], color="lightblue")
sns.boxplot(x=quake['Ms'], ax=axes[7], color="khaki")
sns.boxplot(x=quake['Mb'], ax=axes[8], color="lightgrey")
fig.subplots_adjust(hspace=0.5)
plt.show()
```

Creamos una nueva columna que se llama City_Name

```python
# Define a function to extract city names from location names
def extract_city(location):
    # Check if the location contains parentheses
    if "(" in location and ")" in location:
        # Extract the text within parentheses as the city name
        city_name = location.split("(")[-1].split(")")[0]
        return city_name.strip()  # Remove leading/trailing whitespace
    else:
        return location  # Return the original location if no parentheses found

# Create a new column 'City_Name' by applying the extract_city function to the 'Location' column
quake['City_Name'] = quake['Location'].apply(extract_city)

# Print the DataFrame to verify the results
print(quake)
```

Convertimos carácter turco en carácter ingles

```python
# Convert turkish characters to english characters
# Import the unidecode library
import unidecode

# Convert UTF-8 characters to ASCII in the 'City_Name' column
quake['City_Name'] = quake['City_Name'].apply(unidecode.unidecode)

# Print the DataFrame to verify the results
print(quake)
```

```python
# Count the number of earthquakes in each city
city_counts = quake['City_Name'].value_counts()

print("Top 10 cities with the most earthquakes:")
print(city_counts.head(20))
print()
print(city_counts.tail(20))
```
