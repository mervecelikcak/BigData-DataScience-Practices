```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set an array for question 8
realtime1 = np.array([51.606, 50.935, 49.743, 53.686])
realtime2 = np.array([51.591, 50.770, 50.199, 50.659])
realtime3 = np.array([50.795, 50.515, 50.291, 50.221])
```

```python
# Calculate the mean and standard deviation for each array
mean1 = np.mean(realtime1)
std1 = np.std(realtime1)
mean2 = np.mean(realtime2)
std2 = np.std(realtime2)
mean3 = np.mean(realtime3)
std3 = np.std(realtime3)

# Print the mean and standard deviation for each array with formatting and rounding
print("Mean of realtime1: {:.3f}".format(mean1))
print("Standard deviation of realtime1: {:.3f}".format(std1))
print("Mean for of realtime2: {:.3f}".format(mean2))
print("Standard deviation of realtime2: {:.3f}".format(std2))
print("Mean of realtime3: {:.3f}".format(mean3))
print("Standard deviation of realtime3: {:.3f}".format(std3))

# Print as a table
print("Mean and Standard Deviation for each array:")
print("realtime1: {:.3f} {:.3f}".format(mean1, std1))
print("realtime2: {:.3f} {:.3f}".format(mean2, std2))
print("realtime3: {:.3f} {:.3f}".format(mean3, std3))
```

```python
# Concatenate the arrays into a DataFrame
df = pd.DataFrame({'Realtime1': realtime1, 'Realtime2': realtime2, 'Realtime3': realtime3})

# Plot a boxplot mean and standard deviation for each array
fig, axes = plt.subplots(1, 1, figsize=(10, 3))
# Vertical boxplot
sns.boxplot(data=df, ax=axes, orient='h', width=0.4)
plt.show()
```
